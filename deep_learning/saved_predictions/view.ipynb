```python
import numpy as np
import seaborn as sns

from sklearn.metrics import confusion_matrix, precision_recall_fscore_support, roc_auc_score, roc_curve, matthews_corrcoef, plot_confusion_matrix
from collections import defaultdict
```

```python
cnn_preds = np.load('cnn.npy')
rnn_preds = np.load('rnn.npy')
cnn_lite_preds = np.load('cnn_lite.npy')
roberta_preds = np.load('roberta.npy')
```

```python
y_test = np.load('../../datasets/splits/y_test.npy')
y_test = y_test * 100
```

```python
def compare(y_true, y_preds):
    true_labels = [1 if x > 20 else 0 for x in y_true]
    predictions_labels = [1 if x > 20 else 0 for x in y_preds]
    print('MCC', matthews_corrcoef(true_labels, predictions_labels)  )
    cf_matrix = confusion_matrix(true_labels, predictions_labels)
    sns.heatmap(cf_matrix, annot=True, fmt='g')
```

```python
compare(y_test, cnn_preds)
```

```python
compare(y_test, rnn_preds)
```

```python
compare(y_test, cnn_lite_preds)
```

```python
compare(y_test, roberta_preds)
```
